# What is Exponential Growth & Decay?

A quantity changes at a rate **proportional to its current value**.

### Key Idea:
- Bigger value → faster change
- Smaller value → slower change


# General Formula

$$
y = a e^{kt}
$$

### Terms:

- $y$ → value at time $t$
- $a$ → initial value
- $k$ → growth/decay constant
- $e \approx 2.718$


# Growth vs Decay

### Exponential Growth
- $k > 0$
- Rapid increase

Examples:
- Population growth
- Compound interest

---

### Exponential Decay
- $k < 0$
- Decrease over time

Examples:
- Radioactive decay
- Battery discharge


# Graph Visualization using Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate time values
t = np.linspace(0, 5, 100)

# Growth and decay
growth = np.exp(t)
decay = np.exp(-t)

plt.figure(figsize=(8,5))

plt.plot(t, growth, label='Exponential Growth (e^t)')
plt.plot(t, decay, linestyle='--', label='Exponential Decay (e^-t)')

plt.axhline(0)
plt.axvline(0)

plt.title('Exponential Growth vs Decay')
plt.xlabel('Time (t)')
plt.ylabel('Value')

plt.grid(True)
plt.legend()

plt.show()
```

# Graph Insights

### Growth Curve
- Starts slow
- Increases rapidly
- Explodes over time

### Decay Curve
- Drops quickly initially
- Slows down later
- Approaches zero (never touches)


# Discrete Form

$$y = a(1 + r)^t$$

- $r > 0$ → growth
- $r < 0$ → decay


# Key Concepts

### Doubling Time
$$T = \frac{\ln 2}{k}$$

### Half-Life
$$T = \frac{\ln 2}{|k|}$$


# Example (Growth)

```python
import numpy as np

t = 5
y = 100 * np.exp(0.2 * t)

print('Value after t=5:', y)
```

# Example (Decay)

```python
t = 5
y = 50 * np.exp(-0.1 * t)

print('Value after t=5:', y)
```
