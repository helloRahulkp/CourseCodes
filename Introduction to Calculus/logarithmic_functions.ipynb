# What is a Logarithm?

A logarithm answers:

> To what power should we raise a number to get another number?

### Definition

$$\log_b(a) = x \iff b^x = a$$

- Base = $b$
- Number = $a$
- Result = $x$


### Example

$$\log_2(8) = 3$$

Because:

$$2^3 = 8$$

# Logarithmic Function

$$y = \log_b(x)$$

This is the **inverse** of:

$$y = b^x$$

# Graph of Log Function

### Key Properties

- Domain: $x > 0$
- Range: $(-\infty, \infty)$
- Passes through: $(1, 0)$
- Vertical asymptote: $x = 0$


# Important Log Rules

### Product Rule
$$\log_b(xy) = \log_b(x) + \log_b(y)$$

### Quotient Rule
$$\log_b\left(\frac{x}{y}\right) = \log_b(x) - \log_b(y)$$

### Power Rule
$$\log_b(x^n) = n \cdot \log_b(x)$$

### Change of Base
$$\log_b(x) = \frac{\log(x)}{\log(b)}$$


# Common Log Types

### Natural Log
$$\ln(x) = \log_e(x)$$

$$e \approx 2.718$$

Used in:
- Calculus
- Machine Learning
- Growth/Decay models

---

### Common Log
$$\log(x) = \log_{10}(x)$$

Used in:
- Engineering
- Scientific notation


# Solving Log Problems

### Example 1

$$\log_3(x) = 2$$

Convert:

$$x = 3^2 = 9$$

### Example 2

$$\log(x) + \log(2) = \log(10)$$

Apply product rule:

$$\log(2x) = \log(10)$$

$$2x = 10 \Rightarrow x = 5$$

# Logarithmic Function Visualization using Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (must be > 0 for log)
x = np.linspace(0.1, 10, 100)

# Compute log values
y_log10 = np.log10(x)     # log base 10
y_ln = np.log(x)          # natural log

# Plot
plt.figure(figsize=(8, 5))

plt.plot(x, y_log10, label="log10(x)")
plt.plot(x, y_ln, linestyle='--', label="ln(x)")

# Axes lines
plt.axhline(0)
plt.axvline(0)

# Labels & Title
plt.title("Logarithmic Functions")
plt.xlabel("x")
plt.ylabel("y")

# Grid & Legend
plt.grid(True)
plt.legend()

# Show plot
plt.show()
```
