## 3. Create Tables and Insert Sample Data

```python
%%sql
-- Create Students table
CREATE TABLE students (
    student_id INTEGER PRIMARY KEY,
    first_name VARCHAR(50) NOT NULL,
    last_name VARCHAR(50) NOT NULL,
    email VARCHAR(100) UNIQUE,
    enrollment_date DATE,
    major VARCHAR(50)
);
```

```python
%%sql
-- Create Courses table
CREATE TABLE courses (
    course_id INTEGER PRIMARY KEY,
    course_name VARCHAR(100) NOT NULL,
    credits INTEGER,
    department VARCHAR(50)
);
```

```python
%%sql
-- Create Enrollments table (junction table for many-to-many relationship)
CREATE TABLE enrollments (
    enrollment_id INTEGER PRIMARY KEY,
    student_id INTEGER,
    course_id INTEGER,
    grade VARCHAR(2),
    semester VARCHAR(20),
    FOREIGN KEY (student_id) REFERENCES students(student_id),
    FOREIGN KEY (course_id) REFERENCES courses(course_id)
);
```

## 4. SELECT Queries

### Select All Columns

```python
%%sql
-- Select all students
SELECT * FROM students;
```

### Select Specific Columns

```python
%%sql
-- Select specific columns from students
SELECT name, age FROM students;
```

## 5. WHERE Clause - Filtering Data

```python
%%sql
-- Filter students by major
SELECT * FROM students
WHERE major = 'Computer Science';
```

```python
%%sql
-- Multiple conditions with AND
SELECT * FROM students
WHERE major = 'Computer Science' AND enrollment_date >= '2023-09-01';
```

```python
%%sql
-- OR condition
SELECT * FROM students
WHERE major = 'Mathematics' OR major = 'Physics';
```

```python
%%sql
-- IN operator
SELECT * FROM students
WHERE major IN ('Mathematics', 'Physics', 'Chemistry');
```

```python
%%sql
-- LIKE pattern matching
SELECT * FROM students
WHERE email LIKE '%@university.edu';
```

## 6. ORDER BY - Sorting Results

```python
%%sql
-- Sort by last name ascending (default)
SELECT first_name, last_name, major FROM students
ORDER BY last_name;
```

```python
%%sql
-- Sort by multiple columns
SELECT first_name, last_name, major, enrollment_date FROM students
ORDER BY major ASC, enrollment_date DESC;
```

## 7. GROUP BY - Aggregating Data

```python
%%sql
-- Count students by major
SELECT major, COUNT(*) AS student_count
FROM students
GROUP BY major;
```

```python
%%sql
-- Average credits per course by department
SELECT department, AVG(credits) AS avg_credits, COUNT(*) AS course_count
FROM courses
GROUP BY department;
```

```python
%%sql
-- HAVING clause (filter after grouping)
SELECT major, COUNT(*) AS student_count
FROM students
GROUP BY major
HAVING COUNT(*) > 1;
```

## 8. JOINs - Combining Tables

### INNER JOIN

```python
%%sql
-- Join students with their enrollments
SELECT 
    s.first_name,
    s.last_name,
    c.course_name,
    e.grade,
    e.semester
FROM enrollments e
INNER JOIN students s ON e.student_id = s.student_id
INNER JOIN courses c ON e.course_id = c.course_id;
```

### LEFT JOIN

```python
%%sql
-- All students with their enrollments (including students with no enrollments)
SELECT 
    s.first_name,
    s.last_name,
    e.course_id,
    e.grade
FROM students s
LEFT JOIN enrollments e ON s.student_id = e.student_id;
```

### Multiple JOINs with Aggregation

```python
%%sql
-- Student grades summary
SELECT 
    s.first_name,
    s.last_name,
    COUNT(e.enrollment_id) AS total_courses,
    AVG(CASE 
        WHEN e.grade = 'A' THEN 4.0
        WHEN e.grade = 'A-' THEN 3.7
        WHEN e.grade = 'B+' THEN 3.3
        WHEN e.grade = 'B' THEN 3.0
        WHEN e.grade = 'B-' THEN 2.7
        ELSE 0.0
    END) AS gpa
FROM students s
LEFT JOIN enrollments e ON s.student_id = e.student_id
GROUP BY s.student_id, s.first_name, s.last_name
ORDER BY gpa DESC;
```

## 9. Advanced Queries

```python
%%sql
-- Subquery: Find courses with more than average students enrolled
SELECT 
    c.course_name,
    c.department,
    COUNT(e.student_id) AS enrolled_students
FROM courses c
LEFT JOIN enrollments e ON c.course_id = e.course_id
GROUP BY c.course_id, c.course_name, c.department
HAVING COUNT(e.student_id) > (
    SELECT AVG(enrollment_count)
    FROM (
        SELECT COUNT(student_id) AS enrollment_count
        FROM enrollments
        GROUP BY course_id
    )
);
```
