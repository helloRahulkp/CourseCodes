# Mathematical Operations

## Load SQL Extension

```python
%load_ext sql
```

## Connect to PSQL Database

```python
%sql postgresql://localhost/math_db
```

```python
%%sql
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    score INTEGER
);
```

```python
%%sql
INSERT INTO students VALUES
(1,'Alice',85),
(2,'Bob',92),
(3,'Charlie',67),
(4,'David',74),
(5,'Eva',90);
```

## List all tables

```python
%%sql
SELECT table_name
FROM information_schema.tables
WHERE table_schema = 'public';
```

## Basic Arithmetic

```python
%%sql
SELECT name, score, score + 5 AS boosted_score
FROM students;
```

## Multiple Operations

```python
%%sql
SELECT name, score,
       score * 1.1 AS scaled,
       score / 10 AS normalized
FROM students;
```

## Aggregate Functions

```python
%%sql
SELECT 
AVG(score) AS avg_score,
MAX(score) AS max_score,
MIN(score) AS min_score,
SUM(score) AS total_score,
COUNT(*) AS total_students
FROM students;
```

## Filtering with Math

```python
%%sql
SELECT *
FROM students
WHERE score > (SELECT AVG(score) FROM students);
```

## Mathematical Functions

```python
%%sql
SELECT name, score,
ROUND(score,1),
SQRT(score),
POWER(score,2)
FROM students;
```

# Convert to Pandas

```python
import pandas as pd
result = %sql SELECT * FROM students;
df = result.DataFrame()
df
```

## NumPy Operations

```python
import numpy as np

scores = df['score'].values

scores + 5
scores * 1.1
scores / 10
scores ** 2
```

## Aggregate with NumPy

```python
np.mean(scores)
```

```python
np.max(scores)
```

```python
np.min(scores)
```

```python
np.sum(scores)
```

## Advanced Math

```python
np.sqrt(scores)
np.log(scores)
np.exp(scores)
```

## ML Feature Scaling

```python
normalized = (scores - np.mean(scores)) / np.std(scores)
normalized
```
