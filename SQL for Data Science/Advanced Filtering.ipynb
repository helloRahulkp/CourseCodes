# Create Tables

```python
%%sql
create table employees (
    emp_id int primary key,
    name text,
    department text,
    salary int,
    city text
);
```

# Insert Data

```python
%%sql
insert into employees values
(1, 'Alice', 'Engineering', 70000, 'Delhi'),
(2, 'Bob', 'HR', 50000, 'Mumbai'),
(3, 'Charlie', 'Engineering', 80000, 'Delhi'),
(4, 'David', 'Finance', 60000, 'Chennai'),
(5, 'Eva', 'HR', 55000, 'Bangalore'),
(6, 'Frank', 'Engineering', 90000, 'Mumbai'),
(7, 'George', 'Finance', 65000, 'Delhi');
```

```python
%%sql 
select * from employees;
```

# Convert into Pandas

```python
import pandas as pd
result = %sql select * from employees;
df = result.DataFrame()
df.head(7) 
```

# IN Operator

```python
%%sql
select * from employees where city in ('Delhi','Mumbai');
```

```python
df[df['city'].isin(['Delhi','Mumbai'])]
```

# AND Operator

```python
%%sql
select * from employees where department='Engineering' and salary > 75000;
```

```python
df[(df['department']=='Engineering') & (df['salary']>75000)]
```

# OR Operator

```python
%%sql
select * from employees where department='HR' or salary > 80000;
```

```python
df[(df['department']=='HR') | (df['salary']>80000)]
```

# NOT Operator

```python
%%sql
select * from employees where not department = 'HR'
```

```python
df[df['department']!='HR']
```

# LIKE + Wildcards

```python
%%sql
select * from employees where name like 'A%'
```

```python
df[df['name'].str.startswith('A')]
```
