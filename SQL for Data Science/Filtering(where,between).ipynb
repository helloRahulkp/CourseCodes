## Create Tables

```python
%%sql
create table users (
    user_id SERIAL primary key,
    name text,
    email text,
    city text
);
```

```python
%%sql
create table products (
    product_id SERIAL primary key,
    product_name text,
    price float
);
```

```python
%%sql
create table orders (
    order_id SERIAL primary key,
    user_id int,
    product_id int,
    amount float,
    order_date date,
    foreign key (user_id) references users (user_id),
    foreign key (product_id) references products (product_id)
);
```

## Insert Data

```python
%%sql
insert into users (name, email,city) values
('Alice', 'ali@gmail.com','Delhi'),
('Bob', NULL,'Mumbai'),
('charlie','charlie@gmail.com','Bangalore'),
('David', NULL, 'Delhi');
```

```python
%%sql
insert into products (product_name, price) values
('MacBook', 10000000),
('iPhone',300000),
('iPad', 500000);
```

```python
%%sql
insert into orders (user_id, product_id, amount,order_date) values
(1,1,75000,'2024-01-10'),
(2,2,30000,'2024-02-15'),
(3,3,20000,'2024-03-20');
```

## WHERE

```python
%%sql
select * from users where city='Delhi'; 
```

## BETWEEN

```python
%%sql 
select * from products where price between 20000 and 500000;
```

## NULL

```python
%%sql
select * from users where email is NULL;
```

## Convert to Pandas

```python
import pandas as pd
```

```python
users = %sql select * from users;
users_df = users.DataFrame()
users_df.head() 
```

## Pandas WHERE

```python
users_df[users_df['city']=='Delhi']
```

## Panadas BETWEEN

```python
products = %sql select * from products;
products_df = products.DataFrame()
products_df[products_df['price'].between(20000,500000)]
```

## Pandas NULL

```python
users_df[users_df['email'].isnull()]
```
